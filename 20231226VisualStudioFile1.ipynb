```python
import ee
import geemap
import datetime

# Initialize Earth Engine
ee.Initialize()

# Define the FAO/GAUL administrative units dataset
dataset = ee.FeatureCollection('FAO/GAUL/2015/level2')

# Define a region of interest (ROI) for India
india_roi = ee.Geometry.Polygon(
    [[[
      68.11137880340998, 7.96553484576015],
      [88.3927306205192, 7.96553484576015],
      [88.3927306205192, 35.67459030295267],
      [68.11137880340998, 35.67459030295267],
      [68.11137880340998, 7.96553484576015]
    ]]
)

# Filter the administrative units dataset to select India
india_dataset = dataset.filterBounds(india_roi)

# Define a region of interest (ROI) for New Delhi
delhi_roi = ee.Geometry.Point(77.216721, 28.613939)

# Filter the India dataset to select New Delhi
delhi_feature = india_dataset.filterBounds(delhi_roi)

# Create a Map
Map = geemap.Map()
Map.centerObject(india_roi, 5)  # Center the map on India

# Add a layer for New Delhi
Map.addLayer(delhi_feature, {'color': 'FF0000'}, 'New Delhi')

# Buffer New Delhi and filter surrounding administrative units
buffered_delhi = delhi_feature.geometry().buffer(0.1)  # Adjust the buffer distance as needed
surrounding_admin_units = india_dataset.filterBounds(buffered_delhi)

# Add a layer for surrounding administrative units
Map.addLayer(surrounding_admin_units, {'color': 'Yellow'}, 'Surrounding Administrative Units')

# Function to calculate NDVI for Landsat imagery
def calculate_ndvi(image):
    ndvi = image.normalizedDifference(['B4', 'B3']).rename('NDVI')
    return image.addBands(ndvi)

# Function to calculate NDMI for Landsat imagery
def calculate_ndmi(image):
    ndmi = image.normalizedDifference(['B4', 'B5']).rename('NDMI')
    return image.addBands(ndmi)

# Function to calculate tree canopy cover for Landsat imagery
def calculate_tree_canopy(image):
    # Calculate NDVI
    ndvi = image.normalizedDifference(['B4', 'B3']).rename('NDVI')

    # Define a threshold for NDVI to identify vegetated areas (adjust as needed)
    ndvi_threshold = 0.2

    # Create a binary image where 1 represents vegetated areas
    canopy_cover = ndvi.gt(ndvi_threshold).rename('TreeCanopy')

    return image.addBands(canopy_cover)

# Define the time range for Landsat imagery
start_date = datetime.datetime(2000, 1, 1)
end_date = datetime.datetime(2020, 12, 31)

# Create a list to store imagery for each year
yearly_imagery = []

# Loop through the years (2000, 2005, 2010, 2015, 2020)
for year in range(2000, 2025, 5):
    # Filter Landsat imagery for the current year
    landsat_region = surrounding_admin_units.geometry().union(buffered_delhi)
    
    # Filter Landsat 8 imagery
    landsat8_imagery = ee.ImageCollection("LANDSAT/LC08/C01/T1_SR") \
        .filterBounds(landsat_region) \
        .filterDate(datetime.datetime(year, 1, 1), datetime.datetime(year, 12, 31)) \
        .sort("CLOUD_COVER") \
        .mosaic()
    
    # Filter Landsat 5 imagery as a fallback
    landsat5_imagery = ee.ImageCollection("LANDSAT/LT05/C01/T1_SR") \
        .filterBounds(landsat_region) \
        .filterDate(datetime.datetime(year, 1, 1), datetime.datetime(year, 12, 31)) \
        .sort("CLOUD_COVER") \
        .mosaic()
    
    # Choose Landsat 8 if available, otherwise use Landsat 5
    selected_imagery = ee.ImageCollection(ee.Algorithms.If(
        landsat8_imagery.bandNames().size(),
        landsat8_imagery,
        landsat5_imagery
    ))
    
    # Calculate NDVI for New Delhi and surrounding administrative units
    delhi_with_ndvi = selected_imagery.filterBounds(delhi_feature.geometry()).map(calculate_ndvi)
    surrounding_admin_with_ndvi = selected_imagery.map(calculate_ndvi)

    # Calculate NDMI for New Delhi and surrounding administrative units
    delhi_with_ndmi = selected_imagery.filterBounds(delhi_feature.geometry()).map(calculate_ndmi)
    surrounding_admin_with_ndmi = selected_imagery.map(calculate_ndmi)

    # Calculate tree canopy cover for New Delhi and surrounding administrative units
    delhi_with_canopy = selected_imagery.filterBounds(delhi_feature.geometry()).map(calculate_tree_canopy)
    surrounding_admin_with_canopy = selected_imagery.map(calculate_tree_canopy)

    # Add the NDVI, NDMI, and tree canopy cover layers to the map
    Map.add_layer_control()

# Display the Map
Map
```
