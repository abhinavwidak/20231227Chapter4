```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import textwrap

# Read the CSV file into a DataFrame
data = pd.read_csv("/Users/rishi/3DwlPhilFeb262023/00EcosystemCondition/2EcosystemMatrix_Normalized-1.csv")

# Specify the columns to extract
columns_to_extract = [
    'Parent Region',
    'District',
    'Normalized_2000NDVI',
    'Normalized_2000NDMI',
    'Normalized_2000Tree Canopy Cove',
    'Normalized_2000Green Space Per inhabitant'
]

# Extract the desired columns
df = data[columns_to_extract]

# Get unique parent regions for grouping
unique_parent_regions = df['Parent Region'].unique()

# Define the number of rows and columns for the matrix of spider diagrams
num_rows = len(unique_parent_regions)  # One row per unique parent region
num_columns = len(columns_to_extract[2:])  # One column per specified spider diagram

# Set up a grid of subplots in a matrix form
fig, axes = plt.subplots(num_rows, num_columns, figsize=(15, 10), subplot_kw={'polar': True})

# Flatten the 2D axes array to access each subplot
axes = axes.flatten()

for i, parent_region in enumerate(unique_parent_regions):
    # Filter the DataFrame for the current parent region
    group = df[df['Parent Region'] == parent_region]

    # Define the time periods (concentric circles)
    time_periods = ['2000', '2005', '2010', '2015', '2020']

    # Initialize a dictionary to store the sum of values for each time period
    time_period_sums = {}

    for time_period in time_periods:
        # Initialize a list to store values for the current time period
        values = []

        for column in columns_to_extract[2:]:
            if time_period in column:
                values.extend(group[column].tolist())

        # Calculate the sum of values for the time period
        time_period_sum = sum(values)

        # Store the time period sum in the dictionary
        time_period_sums[time_period] = time_period_sum

    # Create a list of time periods and their corresponding sums
    time_period_names = list(time_period_sums.keys())
    time_period_sums_values = list(time_period_sums.values())

    # Calculate the maximum time period sum, avoiding division by zero
    max_time_period_sum = max(time_period_sums_values) if max(time_period_sums_values) != 0 else 1

    # Normalize the sums to a maximum of 1
    normalized_sums = [sum_val / max_time_period_sum for sum_val in time_period_sums_values]

    # Ensure that the number of labels matches the number of data points
    labels = time_period_names + [time_period_names[0]]

    # Create the spider plot for each parent region
    axes[i].plot(np.deg2rad(np.linspace(0, 360, len(labels))), normalized_sums + [normalized_sums[0]], "o-", linewidth=2)
    axes[i].fill(np.deg2rad(np.linspace(0, 360, len(labels))), normalized_sums + [normalized_sums[0]], alpha=0.25)

    # Set the title to display the parent region name
    axes[i].set_title(parent_region, fontsize=12)

# Remove any unused subplots
for i in range(len(unique_parent_regions) * num_columns, num_rows * num_columns):
    fig.delaxes(axes[i])

# Add a common title for the entire plot
plt.suptitle("Spider Diagrams of Normalized Ecosystem Data by Parent Region", y=1.08)

# Adjust subplot layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Show the subplots
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import textwrap

# Read the CSV file into a DataFrame
data = pd.read_csv("/Users/rishi/3DwlPhilFeb262023/00EcosystemCondition/2EcosystemMatrix_Normalized-1.csv")

# Extract the desired columns
columns_to_extract = [
    'Parent Region',
    'District',
    'EC_2000',
    'EC_2005',
    'EC_2010',
    'EC_2015',
    'EC_2020'
]
df = data[columns_to_extract]

# Define the years for which you want to create spider diagrams
years = ['2000', '2005', '2010', '2015', '2020']

# Define the number of rows and columns for the matrix
num_rows = len(years) // 2  # Assuming you want 2 spider diagrams per row
num_columns = 2

# Set up a grid of subplots in a matrix form
fig, axes = plt.subplots(num_rows, num_columns, figsize=(15, 10), subplot_kw={'polar': True})

# Flatten the 2D axes array to access each subplot
axes = axes.flatten()

for i, year in enumerate(years):
    column_name = f'EC_{year}'
    
    for parent_region, group in df.groupby('Parent Region'):
        values = group[column_name].tolist()
        num_values = len(values)

        angles = np.linspace(0, 2 * np.pi, num_values, endpoint=False).tolist()
        angles += angles[:1]

        labels = [textwrap.fill(label, 10) for label in group['Parent Region'].tolist()]

        axes[i].plot(angles, values + [values[0]], "o-", linewidth=2, label=parent_region)
        axes[i].fill(angles, values + [values[0]], alpha=0.25)

        axes[i].set_xticks(angles[:-1])
        axes[i].set_xticklabels(labels, fontsize=8, rotation=45)
        axes[i].set_title(f'EC_{year}')

        # Add labels for the edges
        for j, angle in enumerate(angles[:-1]):
            ax = axes[i]
            value = values[j]
            label = labels[j]
            ax.annotate(f'{label}: {value}', (angle, value), xytext=(0, 0), textcoords='offset points')

# Remove any unused subplots
for i in range(len(years), num_rows * num_columns):
    fig.delaxes(axes[i])

# Add a legend
plt.legend(loc='upper right', bbox_to_anchor=(1.5, 1.0))

# Show the subplots
plt.suptitle("Ecosystem Condition Spider Diagrams by Year", y=1.08)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import ee
import geemap

# Initialize Earth Engine
ee.Initialize()

# Load the LANDSAT image collection
dataset = ee.ImageCollection('LANDSAT/LT05/C02/T1_TOA') \
    .filterDate('2001-01-01', '2001-12-31')

# Select bands for true color representation
trueColor321 = dataset.select(['B3', 'B2', 'B1'])

# Visualization parameters
trueColor321Vis = {
    'min': 0.0,
    'max': 0.4,
    'gamma': 1.2
}

# Create a Map centered on the Philippines
Map = geemap.Map(center=[13.4000, 122.5621], zoom=6)

# Load administrative boundary feature for the Philippines
study_area = ee.FeatureCollection('FAO/GAUL/2015/level2') \
    .filter(ee.Filter.eq('ADM0_NAME', 'Philippines'))

# Get the geometry of the study area
study_area_geom = study_area.geometry()

# Center the map on the study area
Map.centerObject(study_area_geom, 7)

# Add the study area layer to the map
Map.addLayer(study_area_geom, {}, 'Study Area')

# Add the true color layer to the map
Map.addLayer(trueColor321, trueColor321Vis, 'True Color (321)')

# Loop through the years of interest and add NDVI layers
years = [2001, 2005, 2010, 2015, 2020]  # Add more years if needed
for year in years:
    ndvi = dataset \
        .filterDate(f'{year}-01-01', f'{year}-12-31') \
        .select(['B4', 'B3']) \
        .map(lambda image: image.normalizedDifference(['B4', 'B3']))
    
    ndvi_mean = ndvi.mean()
    
    ndviVis = {
        'min': -1,
        'max': 1,
        'palette': ['blue', 'white', 'green']
    }
    
    # Add NDVI layer to the map
    Map.addLayer(ndvi_mean, ndviVis, f'NDVI {year}')

# Display the map with layer control
Map.addLayerControl()
Map
```

## Export file with NDVI Values
    These values are the mean of ndvi for all pixels within the administrative region. 

```python

import ee
import geemap

# Initialize Earth Engine
ee.Initialize()

# Load the administrative boundaries (ADM2 districts)
adm2 = ee.FeatureCollection("FAO/GAUL/2015/level2")

# Define the study area
study_area = adm2.filter(ee.Filter.eq('ADM0_NAME', 'Philippines'))

# Define the NDVI collection and time range
ndvi_collection = ee.ImageCollection("MODIS/006/MOD13A1").select('NDVI')
start_date = '2001-01-01'
end_date = '2001-12-31'

# Filter the NDVI collection by date and study area
ndvi_filtered = ndvi_collection.filterDate(start_date, end_date).filterBounds(study_area)

# Calculate the average NDVI for each district
def calculate_average_ndvi(feature):
    geom = feature.geometry()
    mean_ndvi = ndvi_filtered.mean().reduceRegion(
        reducer=ee.Reducer.mean(),
        geometry=geom,
        scale=250
    )
    return feature.set(mean_ndvi)

study_area_with_ndvi = study_area.map(calculate_average_ndvi)

# Convert Earth Engine result to client-side Python object
ndvi_list = study_area_with_ndvi.getInfo().get('features')

# Write results to a CSV file
csv_file = 'average_ndvi.csv'
with open(csv_file, 'w') as f:
    f.write('ADM2_NAME,mean\n')
    for feature in ndvi_list:
        properties = feature.get('properties')
        adm1_name = properties.get('ADM1_NAME')
        adm2_name = properties.get('ADM2_NAME')
        mean_ndvi = properties.get('NDVI_mean')
        f.write(f'{adm1_name},{adm2_name},{mean_ndvi}\n')

print(f'Results exported to {csv_file}')
```

```python
import ee
import geemap
import pandas as pd

# Initialize Earth Engine

ee.Initialize()

# Load district boundaries from FAO GAUL dataset (Level 2)

# Load district boundaries from FAO GAUL dataset (Level 2)

district_boundaries = ee.FeatureCollection('FAO/GAUL/2015/level2').set('crs', 'EPSG:4326')


# Define the time periods

years = [2001, 2005, 2010, 2015, 2020]
start_dates = [f'{year}-01-01' for year in years]
end_dates = [f'{year}-12-31' for year in years]

# Define a function to calculate the mean value for a region

# Define a function to calculate the mean value for a region

def calculate_mean(image, geometry):
    mean_value = image.reduceRegion(
        reducer=ee.Reducer.mean(),
        geometry=geometry,
        scale=500,  # Adjust scale as needed
        maxPixels=1e13,
        crs='EPSG:4326'  # Use the CRS of your data
    )
    return ee.Feature(geometry, mean_value)

# Initialize an empty list to store the results

results = []

# Loop through the years and calculate mean values

for i, (start_date, end_date) in enumerate(zip(start_dates, end_dates)):
    modis_lc = ee.ImageCollection("MODIS/006/MCD12Q1").filterDate(start_date, end_date)
    modis_lc_clip = modis_lc.map(lambda img: img.clip(district_boundaries))

    for lc in range(1, 18):
        lc_image = modis_lc_clip.select('LC_Type1').filter(ee.Filter.eq('LC_Type1', lc)).mean()
        lc_mean = lc_image.reduceRegions(
            collection=district_boundaries,
            reducer=ee.Reducer.mean(),
            scale=250,  # Adjust scale as needed
            tileScale=4  # Adjust tileScale as needed
        )
        lc_mean = lc_mean.map(lambda feat: feat.set('Year', years[i], 'Landcover', lc))
        results.append(lc_mean)

# Merge the results
result_collection = ee.FeatureCollection(results)

# Set the CRS for each feature in the collection
result_collection_with_crs = result_collection.map(lambda feature: feature.set('crs', 'EPSG:4326'))

# Flatten the collection
flattened_collection = result_collection_with_crs.flatten()

result_df = geemap.ee_to_pandas(flattened_collection).map(lambda feature: feature.set('crs', 'EPSG:4326'))
# Print the results
print(result_df)

```
