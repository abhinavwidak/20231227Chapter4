Chapter 3. Assessing the Impacts of Land-Cover Changes on Ecosystem Services in the Philippines 
 
	Introduction
There are many facets to environmental science, but one of the main ones is the pursuit of new knowledge through exploring the complexities of the natural world and the effects of human activity on ecosystems. The use of this most recent knowledge, which improves its interaction with other disciplines to promote a more thorough integration, is another essential component. By allowing for the adaptation and use of environmental insights in several contexts, for a range of goals, and at different points in time, this integration helps to expand the scope of its utilization .
With this paper, our goal is to extend the discourse on urbanization beyond mere physical alterations in ecosystems to a deeper understanding of the implications for human life and well-being. Urbanization has reshaped spatial and ecological dynamics globally. As cities expand, the relationship between resource availability and utilization shifts, impacting the quality of ecosystem services provided to both urban and non-urban populations. Ecosystem services (ES) are the benefits humans derive from ecosystems, including clean air, water, and food production. While the consumption of ecosystem services is not limited to neighboring regions, peri-urban areas play a significant role in supplying these services.  In the Philippines, the peri-urban landscape of Tagaytay Highlands serves as a compelling example of the vital role peri-urban areas play in supplying ecosystem services. Located on the outskirts of Metro Manila, Tagaytay Highlands has evolved into a tourism hub, attracting urbanites seeking respite amidst its scenic hills. While visitors enjoy recreational benefits, these peri-urban areas quietly contribute crucial ecosystem services, maintaining air quality and acting as natural water catchments. While Tagaytay hills offer only one example of a landcover outside the city, it suggests how crucial peri-urban areas are in giving urban inhabitants access to necessary resources. Understanding the broader sustainability of urbanization and the dynamics of various urban ecosystems depends on acknowledging the importance of such and more diverse peri-urban areas in providing ecosystem services.  
Ecosystems are complex and interconnected networks of living organisms, and their physical environments play a central role in providing ecosystem services (De Groot et al., 2002). Ecosystems encompass a wide range of natural processes and interactions, including nutrient cycling, biodiversity maintenance, and climate regulation (De Groot et al. 2002). Natural resources such as rivers, groundwater, and soil serve as the cornerstone of ecosystem services because they provide essential raw materials for ecosystem functionality. Similar to land cover, ecosystem services are also prone to long-term changes because of urbanization. There is an interdependency between resources and the supply of ecosystem services. The availability and sustainable utilization of resources directly shapes the provision of critical services for human well-being. Land cover change, often accompanied by industrialization and urban growth, disrupts ecosystems within and around cities, introducing newer and more difficult challenges for achieving urban sustainability. 
It is difficult to assess changes in ES supply in a rapidly transforming ecosystem, particularly in regions characterized by frequent land cover alterations. Cities have evolved from being the sources of resource extraction to complex hubs where resources, including water, are transported from different surrounding regions to meet the demands of the large and dense population. Such a transition necessitates a comprehensive understanding of the consequences of urbanization on ecosystems (Richards & VanWey, 2015). While urbanization may not rely on local ecosystems for resources in the same way as historical agricultural or extractive activities, it exerts a considerable impact on the environment. The multifaceted nature of ecosystems, marked by their interdependency and diversity, makes it challenging to gauge the changes stemming from land cover transformations. 
The complexity of ecosystems and the dynamic nature of urban land cover change poses a challenge to sustainable development anywhere in the world. Traditional ecological assessments particularly the ones mentioned in Millennium Ecosystem Assessment (MEA) report (MEA, 2005, p. 5), Economics of Ecosystems and Biodiversity (TEEB) report, and the new Intergovernmental Science-Policy Platform on Biodiversity and Ecosystem Services (IPBES) have all taken up identification, valuation, and preservation of ES as key concerns. Despite ample attention paid to defining ES, major challenges remain in operationalizing the concept. These challenges include identifying what counts as an ecosystem service, given that discussions range from physical goods such as timber to intangible cultural services (Fisher et al., 2009) how these services could be valued either economically or non monetarily(Davidson, 2013), and how payments, compensation, or exchange for conservation or provision of ES might be designed and delivered, where appropriate(Oraphan & Jirawan, 2019). Because of their resource limits, many current methodologies, which rely on field research and chemical assessments of soil and water quality, frequently do not provide a comprehensive view of ecosystem health over time. Such a change cannot be quantified quickly enough to reflect the fast transforming urban landcover. The vast variety of land cover types in these environments, each with distinct traits and effects, makes it even more difficult to monitor and assess ecosystem conditions. 
 Understanding the vast array of benefits that ecosystems provide to humans, known as ecosystem services (ES), is critical for both our survival and well-being. These services range from the concrete, like food provision and water filtration, to the more intangible, such as recreational spaces and cultural heritage. However, addressing the challenges in maintaining and enhancing these services is inherently complex. One promising strategy is employing methodologies that quantify how ecosystem changes impact human well-being, which helps us interpret what humans receive from ecosystems as their "supply" of services. Notably, this perceived supply can differ dramatically based on numerous factors  geographic, cultural, and demographic, such as gender, highlighting the diversity of human-ecosystem interactions.
In an effort to standardize the understanding and assessment of ecosystem services, the Common International Classification of Ecosystem Services (CICES) was developed. The CICES provides a framework for categorizing and valuing the plethora of services provided by ecosystems, which is crucial for examining their role in human life (CICES, 2019). With CICES as the backdrop, current techniques aim to measure the supply of ecosystem services by examining how individuals  the ultimate beneficiaries  experience environmental changes. This experiential data is presumed to be indicative of the level of ecosystem service supply. Such an approach not only encompasses the quantifiable benefits like crop yield and clean air but also the non-material advantages that contribute to the quality of life. By drawing on data from those who directly receive these benefits as base, this framework proves instrumental in assessing the impact of urban expansion and land cover changes on ecosystem health across various timespans. 

Literature Review 
The peri-urban interface, a transitional zone blending urban and rural characteristics, is increasingly recognized as a critical area for sustainable development. This zone is often the first to experience the pressures of urban sprawl, leading to rapid and sometimes haphazard growth. The consequences of such expansion are multifaceted, affecting the ecosystem services that are vital to human well-being. These services encompass a wide range of benefits, including provisioning services like food and water; regulating services that affect climate, floods, and disease; supporting services such as soil formation and nutrient cycling; and cultural services that provide recreational, aesthetic, and spiritual benefits (MEA Millennium Ecosystem Assessment (MA) & Synthesis, 2005). As urban areas continue to grow, the challenge is to manage this growth in a way that preserves the ecosystem services provided by peri-urban areas. 
We will now analyze existing studies on land cover change and ecosystem services in peri-urban areas. The unchecked urban expansion into peri-urban zones can lead to significant environmental degradation. Urban sprawl often results in habitat fragmentation and loss, which can negatively impact biodiversity, including the richness and abundance of species (Beninde et al., 2015). The conversion of land from natural or agricultural uses to urban or industrial uses can also result in the loss of valuable ecosystem services. For instance, the degradation of water resources due to urbanization can lead to a reduction in both the quality and quantity of water available for human consumption, agriculture, and other uses (McDonald et al., 2008). Additionally, the loss of green spaces and biodiversity in these areas can lead to diminished air and water quality, increased vulnerability to climate change, and a loss of recreational spaces for urban populations. 
The literature also delves into the opportunities that lie if we manage the interaction between land cover and ecosystem services more carefully. The concept of green infrastructure, which includes networks of natural and semi-natural areas, has been proposed as a way to maintain and enhance ecosystem services while accommodating urban growth (Lafortezza et al., 2013). Green infrastructure can provide a multitude of services, such as stormwater management, air quality improvement, and mitigation of urban heat island effects (Benedict & McMahon, 2012). They form a buffer zone and in many cases support global biodiversity by connecting protected natural areas. This can be especially important for species migration and ecological processes(Livesley et al., 2016). 
In addition to environmental benefits, peri-urban areas can also support sustainable agricultural practices that contribute to urban food security. As cities expand, peri-urban agriculture can provide fresh produce to urban centers, while also offering ecosystem services like pollination, pest control, and maintenance of soil fertility(Zasada et al., 2011). This form of agriculture can be particularly important in developing countries, where food security is a pressing concern. However, to realize the potential of peri-urban areas in providing ecosystem services, there must be a concerted effort to implement sustainable land-use practices, protect green spaces, and promote biodiversity. Integrated urban-rural planning, the preservation of green spaces, sustainable agricultural practices, and the protection of biodiversity are essential strategies for maintaining the ecosystem services upon which both urban and peri-urban populations depend (Lasco et al., 2014). As urban areas continue to grow, the importance of peri-urban zones in achieving sustainable development becomes increasingly clear, necessitating innovative approaches to urban planning that balance development with ecological conservation. A vast body of research on land cover change uses high resolution images to extract and analyze latest data. Most of these studies which use satellite data for analysis account for a mismatch of scale. For example, comparing a 30m resolution imagery with finer images of 10 m resolution would require statistical transformation, or categorical reordering. Addressing the difference between spatial scales of ecosystem condition and land cover data is crucial for a detailed and multifaceted analysis of the relationship between land cover changes and ecosystem service provision(Vindigni et al., 2021). Chen et al. (2023) underline the necessity of integrating both time and space in research to comprehend the trends in ecosystem health and service delivery. Complementing this perspective, Githiora-Murimi et al. (2022) advocate for involvement with local ecosystems to understand the full spectrum of peri-urban ecosystem dynamics.
The unavoidable incongruity between the spatial resolutions of ecosystem condition and land cover data necessitates sophisticated analytical strategies. Recognizing this spatial discord, several studies advocate for methodological adaptations, including the downscaling of coarse-resolution data, statistical harmonization, or the deployment of multiscale indicators to bridge this gap (Burkhard et al., 2013; Dworczyk & Burkhard, 2022; Troy & Wilson, 2006). The Common International Classification of Ecosystem Services (CICES), which has been continually refined following its precursors such as the Millennium Ecosystem Assessment and The Economics of Ecosystems and Biodiversity (TEEB), offers a structured approach with multiscale applicability. This classification system has matured into a robust framework, facilitating the consistent and comprehensive categorization of ecosystem services across various scales and studies(Haines-Young & Potschin, 2012).The adaptability of CICES, allowing for its application across a spectrum from broad Sections to detailed Class Types, aligns with the diverse needs of ecosystem service assessments, a versatility also explored in the recent study by Dworczyk & Burkhard (2022).
Metro Manila serves as the epicenter of our analysis for this study since it is an area that is rapidly experiencing urban sprawl and the ecological consequences that follow. The study explores Metro Manila's urban expansion from 2000 to 2020 and makes the case for a careful examination of the ecological effects of such development. In context of the Philippines, rapid expansion of urban areas is significantly linked to effects such as the degradation of green spaces, heightened urban heat island impact, and encroachment on critical natural resources. The Laguna de Bay region, a significant ecological resource situated in peri-urban Metro Manila, serves as an example of this phenomena of progressive ecological degradation. It is a feature of great economic significance because it promotes rice farming, fisheries, and other economic activities that provide food, employment, and other necessities for a smooth operation of the metropolis (Nor et al., 2017; Tiangco et al., 2008; Iizuka et al., 2017). Beyond the aesthetic concerns, this loss diminishes the ecological services crucial for biodiversity, pollution reduction, and the general quality of life. The observed inverse relationship between the normalized difference vegetation index and temperature underscores the pivotal role of urban greenery in mitigating the urban heat island effect, exacerbating energy consumption and contributing to global warming (Tiangco et al., 2008). Predictive modeling of urban sprawl around the Laguna de Bay reveals potential landscape changes threatening hydrological stability, with profound implications for both biodiversity and human communities (Iizuka et al., 2017). The study projected that an increase in built-up areas emphasized an urgent need to align urban growth with environmental sustainability. Simulations highlighting gaps between planned and actual urban development unveil a disconnect between urban policy and implementation, suggesting the need for new frameworks (Nor et al., 2017). These frameworks should ensure congruence between urban growth and environmental and social priorities, bridging the divide revealed by the disparities between policy intent and real-world outcomes..

	Methodology
	Landcover Change
Land cover change and its impacts have been studied through different approaches in the present literature. Many studies use data they gather and analyze through Remote Sensing and GIS (Ajmal & Jamal, 2021; Biswas & Ghosh, 2021). By analyzing satellite data using geographic information systems (GIS), researchers can quantify the extent and patterns of urban expansion, as well as its impact on peri-urban ecosystems. In another approach, landscape metrics are quantitative measures that assess the spatial characteristics of land cover patterns(Cai et al., 2016; Hesselbarth et al., 2019). Metrics such as fragmentation, connectivity, and diversity can be calculated to evaluate the ecological consequences of urban land cover change. Another valuation method which is often used to calculate the impact of land cover change is the Economic Valuation method(Castillo-Eguskitza et al., 2019; Davidson, 2013). Economic valuation methods assign monetary values to ecosystem services, providing a means to compare their importance with other economic activities. Valuation techniques include market-based approaches, stated preference surveys, and cost-based methods. However, in places where data and resources are scarce, it might be beneficial to adopt the approach which our study also borrows heavily from(Vrebos et al., 2015; Wangai et al., 2019). 
Ecosystem Service Mapping involves identifying and quantifying the supply and demand of ecosystem services in each area. By overlaying land cover data with information on ecosystem functions, we can assess the impact of urbanization on the provision of ecosystem services of peri urban Metro Manila. Many recent papers which use the matrix approach to estimate the change in ecosystem service, rely on SPOT5, Sentinel-2 data (Burkhard et al., 2015; Hattam et al., 2021), or Landsat data (Estoque et al., 2018). All these data sources have a better resolution than MODIS MCD12Q1 (500m) database which this study uses. It uses MODIS dataset to maintain consistency of data across years and to keep the landcover categories broad, yet accurate. Another advantage of this dataset is that it covers the land area for the entire study area for the whole study period in a consistent resolution.

	Calculation of Ecosystem Condition Indicator
Ecosystem Condition Indicator for a district, is the sum of normalized values of:  NDVI, NDMI, Tree Canopy Cover, Green Space Per Inhabitant, Semi-Natural area, Imperviousness per Inhabitant and PM2.5 Equation 1(b). The indicators that make up Ecosystem Condition together show subtle changes in the environment that are not otherwise noticeable. They provide a precise, current, and comprehensible metric that makes cross-temporal and cross-spatial comparison easier. The statistical transformation of absolute values of indicators to their normalized values is shown through Equation 1(a).

ECI=(V-VL)/(VH-VL)      

V is the measured/observed value of the variable,
VH is the high condition value for the variable (upper reference level),
VL is the low condition value (lower reference level).

EC = NDVI (Normalized Difference Vegetation Index) + NDMI (Normalized Difference Moisture Index) + Green Space per inhabitant (area/unit) + Semi Riparian Landcover (percentage) + Imperviousness per inhabitant (area/unit) + PM2.5 (concentration)
Equation 1(b)- Components of Ecosystem Condition Indicator


These indicators can be classified into four broad categories: physical, chemical, structural, and functional. An example of the indicators used for assessing the physical state is imperviousness per inhabitant, which indicates the extent of built-up areas and the associated change in permeable surfaces over time. The assessment of air pollutant concentrations helped evaluate the chemical state of the ecosystem, providing information on air quality and potential impacts on human health. The consideration of structural indicators such as greenness, tree canopy cover, and the extent of green spaces offered insights into the overall structure and functioning of urban ecosystems. These indicators shed light on the availability of green areas, tree cover, and the potential for ecosystem services such as temperature regulation, carbon sequestration, and recreational opportunities(de Groot et al., 2002). Furthermore, the analysis of woody vegetation  and semi natural riparian cover was assumed to be a proxy for reflecting the diversity and abundance of plant and animal species in the urban landscape indicates the functional characteristics of ecosystem(Dittrich et al., 2017).  To assess change in the physical state of peri urban ecosystems, we divided total urban landcover by population, and the Normalized Difference Moisture Index (NDMI) derived from LANDSAT data. The assessment of the chemical state was carried out by analyzing air pollutant concentrations obtained from the annual AQ Statistics at the district level. 
To effectively calculate Ecosystem Condition Indicators, we require a clear reference point against which we can evaluate the current condition. Traditionally, this reference has been based on either a historical baseline representing an undisturbed ecosystem condition or the pristine state of the ecosystem(Hatziiordanou et al., 2019). However, in some cases, such as highly urbanized areas like NCR, finding pristine or historical references can be challenging. In such instances, a different approach is usually adopted to establish the reference point(Haase et al., 2014; Troy & Wilson, 2006). In our specific case, we have chosen a reference year of 2000. This decision is based on the readily available empirical data from the past 20 years, as well as the rapid changes in the urban environment observed in these districts during this time frame. By examining the database encompassing this time frame, we identify the maximum and minimum values recorded for each ecosystem condition variable for the whole country. These extreme values serve as the reference points against which the current ecosystem condition can be assessed.

	Ecosystem Matrix 
The method we adopt to assess changes ecosystem service supply over time is the Ecosystem capacity matrix approach (Burkhard et al., 2015; Hattam et al., 2021). In this method the capacity or potential of an ecosystem service providing unit (typically landuse or habitat types) to supply ecosystem services is assessed and scored. The method of scoring can draw on multiple data sources (e.g. literature, model output, observational data) and where data is scarce, expert judgement and knowledge is incorporated(Campagne & Roche, 2018). This study utilizes an ecosystem service potential matrix developed by Hattam et al. (2021) to assess habitat contributions to services.
Hattam et al. (2021) applied the ecosystem capacity matrix approach to capture relationships between habitats and services clearly. They defined 13 macro and 21 micro habitats and 18 key services for Southeast Asia based on literature and stakeholder consultation. An extensive review then gathered evidence of links between each habitat-service combination, supplemented by expert knowledge where data was limited. Country teams initially scored matrices individually based on 2021). Confidence levels accompanied scores to acknowledge uncertainties. Matrices and evidence underwent cross-checks and peer review for validation. This standardized yet adaptable methodology allowed Hattam et al. (2021) to develop a matrix applicable across data-limited Southeast Asian marine regions. Habitats included natural, modified, intertidal and subtidal classes relevant to case study sites. The resulting matrix and evidence base form a valuable baseline for ecosystem service assessments in the region (Hattam et al., 2021). This study utilizes the matrix scores to assess habitat contributions to key services in the study area.
Once the ecosystem service potential matrix is prepared, the next step is to apply the matrix scores spatially. Land cover maps of the study area are overlaid with the matrix scores to visualize potential patterns of ecosystem service delivery across the seascape.

Metro Manila's urban area expanded due to expansion into surrounding areas, and provinces like Batangas, Laguna, Pampanga, Bulacan, and Cavite also experienced significant urban expansion. Agricultural trends showed that Isabela maintained high crop outputs through intensification and irrigation. Kalinga's croplands expanded with improvements in road access, benefiting highland vegetable and rice farms. On the other hand, Pangasinan experienced cropland losses due to urbanization, conversion to built-up areas, and infrastructure projects. Forest losses were more prominent near urban zones compared to remote areas.
In terms of land cover composition, urban land cover increased from 2258 sq km in 2001 to 2371 
sq km in 2020, while non-urban land cover decreased slightly from 138619 sq km to 136097 sq km. Grasslands generally benefited from recovery on abandoned farmlands.

Discussion
Our investigation into land cover dynamics within and in the vicinity of Metro Manila has provided empirical evidence of grassland and forest depletion concomitant with urban expansion. This finding aligns with prior studies that documented habitat loss and fragmentation due to urban sprawl in the region. Our findings also indicate a potential linkage between deteriorating ecosystem conditions and land cover transformations in densely urbanized locales such as Batangas and Laguna, reinforcing the notion that urban growth may be exerting adverse effects on peri-urban environmental quality over time.
The data suggests that the total expanse of agricultural land was greater in 2020 compared to 2000. The expansion of croplands and the persistence of grasslands in certain districts point towards agricultural intensification and ecological recovery on lands previously abandoned from farming practices, which may be mitigating some ecological consequences. By normalizing the values of ecosystem service supply, the disproportionate contribution of areas like Zambales and Quezon to Metro Manila's overall ecosystem service provision became evident. A more detailed inquiry could pinpoint crucial areas for conservation within these service-providing zones.
The heavy dependence of Metro Manila on adjacent regions for essential provisioning services, including air and water purification, suggests a paucity of natural landscapes or permeable surfaces within the city to facilitate groundwater recharge. It also implies that the land cover of bordering peri-urban zones plays a pivotal role in shaping the direction and extent of urban sprawl. These observations lend support to the argument that the land cover alterations prompted by urbanization are transforming ecosystem functions in ways that could be harmful.
Due to its coarseness, the resolution of data employed in this study merits cautious interpretation, as it may lead to an underestimation of the localized impacts on ecosystems. Further research is imperative to validate the current findings, to explore more landscapes and habitats and to deconstruct landscape patterns with enhanced detail. Advancing the matrix approach even within the Common International Classification of Ecosystem Services (CICES) framework would involve moving towards a more detailed classification and assessment strategy. This refinement requires the implementation of extensive field surveys designed to capture high-resolution, location-specific data. Such data collection efforts would allow for a finer analysis of ecosystem service provision, potentially revealing subtle changes and trends that broader remote sensing assessments like the one used in this paper may fail to capture.

# Urban Expansion and Ecosystem Health: Exploring the Dynamics of Land-Cover Changes and Ecosystem Services in the National Capital Region of India

## Introduction
Urbanization, with its ever-expanding footprint, demands a critical examination in the context of sustainable urban development. In this regard, the exploration of land-cover changes and their ramifications for Ecosystem Services (ES) and Ecosystem Condition (EC) holds significant importance (Berton et al., 2017; Feddern et al., 2023; Jonas et al., 2015; Ratcliffe et al., 2016; Vogel et al., 2021). ES encompasses the invaluable benefits provided by ecosystems to humanity, including clean air, water, food production, and waste disposal. EC, conversely, signifies the health and resilience of ecosystems (De Groot et al., 2002).

An essential assumption relevant to the study of urbanization and land-cover changes is the potential scenario where the supply of ES within cities exceeds its own capacity to provide them, leading to an "Ecosystem Service deficit" (Vogel et al., 2021). This phenomenon has profound implications for the sustainability and resilience of urban and peri-urban regions (Feddern et al., 2023).

Our investigation, inspired by Tang, J.M. et al.'s methodological approach (in the JOURNAL OF APPLIED REMOTE SENSING), is predicated on the need to understand the spatial-temporal patterns of urbanization within Delhi's National Capital Region (NCR). This region has been segmented into various administrative units for the purpose of detailed analysis. These include the core urban area of the Delhi metropolis and its fringe satellite cities and towns such as Gurgaon, Noida, Faridabad, and Ghaziabad. Gurgaon, known for its booming IT and finance sectors, has emerged as a major hub for multinational corporations. Its skyline, dotted with skyscrapers, is a testament to its rapid development and modern urban planning. Noida, another significant satellite city, is recognized for its robust infrastructure and is a preferred locale for IT and educational institutions. Faridabad, an industrial stronghold, provides a blend of manufacturing units and residential areas, contributing to the diverse LULC patterns in the region. Ghaziabad, with its strategic location and connectivity, is fast becoming a nexus for residential and commercial development.
By focusing on these units, we seek to understand the ecological implications of rapid urban development. The selection of these units is particularly relevant, as they are representative of the broader trends of land conversion, which is a cornerstone of urban growth and environmental change in the region. Each satellite town, with its unique development trajectory and socio-economic role, informs the composite picture of the NCR's transformation and presents a microcosm of the challenges and opportunities inherent in managing urban sprawl and sustainability. 

## Study Area:
Elaborating on the contextual framework provided by Budhiraja, B. et al. in URBAN CLIMATE, the Delhi NCR is a prime example of heterogeneity in urban form and function. The region is geographically diverse, spanning across the flat terrain of the Gangetic plain and reaching into the hilly contours of the Aravallis. This regional complexity is mirrored in its administrative and governance structure. Various authorities, including municipal corporations, development authorities, and village panchayats, coalesce under the overarching strategy laid out by the NCR Planning Board. The urban tapestry of the NCR, which Budhiraja et al. have parsed through in their thermal performance analysis, is composed of densely populated urban cores, rapidly developing suburban nodes, and extensive rural outskirts. Together, these areas form a mosaic of land use patterns that shape the region's climate, with particular reference to the urban heat island phenomenon. These sub-cities within the NCR each display distinct thermal signatures that reflect their urban design and development policies, which are critical factors in urban climatology and sustainability studies.

## Literature review

#### Peri-urban LULC change and drivers in Delhi NCR from 1990-2018

A number of studies have analyzed land use/land cover change and ecosystem services valuation in the Delhi NCR region of India over the past few decades. Several used remote sensing and geospatial techniques to map changes in landscape patterns due to urbanization from 1990-2016. A significant increase in fragmented built-up areas was observed, with conversion of farmland to impervious surfaces (Balha et al., 2020; Sharma & Joshi, 2016).

Other studies examined how urban growth has impacted environmental factors. Results showed increases in land surface temperature (LST), built-up areas, and indices of density/non-vegetation coinciding with decreases in vegetation (NDVI) and water content (NDWI) from 1998-2011 (Malik et al., 2019; Sharma & Joshi 2016). The spatial distribution and seasonal variations of LST in relation to landscape characteristics were also evaluated (Srivastava & Satyaprakash 2020; Mallick et al., 2020; Mallick 2021).

Social aspects like perceptions affecting healthy food consumption, and impacts of increased temperatures and pollution on life quality, were included (Misra & Singh 2016; Sharma et al., 2020). Migration, tertiary employment, and infrastructure/industrial development emerged as key drivers of changes from 1990-2023 (Naikoo et al., 2020, 2022, 2023; Kumar & Sharma 2023).

Modeling studies estimated built-up expansion probabilities and implications for resource planning (Naikoo et al., 2023). Other assessments evaluated ground subsidence, afforestation effects on heat islands, and water conservation needs (Malik et al., 2019; Srivastava & Satyaprakash 2020; Koundanya & Gupta 2011).

Beyond Delhi NCR, ecosystem services valuation and dynamics linked to land use were estimated across regions using various tools and approaches like InVEST (Tripathi et al., 2019; Talukdar et al., 2020; Saha et al., 2022; Das et al., 2019; Ghosh et al., 2021). Reviews identified declining trends for services in western Himalayas (Mondal & Zhang, 2018).

Ecosystem services, as demonstrated in many studies, play a pivotal role in sustaining human well-being, and their provision is intrinsically linked to land cover dynamics, particularly in rapidly urbanizing areas such as the National Capital Territory (NCT) of Delhi. Peri-urban areas, situated on the urban-rural interface, provide a crucial supporting service by hosting a diverse array of pollinators essential for agriculture. However, urbanization in these regions leads to habitat loss and dwindling pollinator populations, posing a significant threat to food security (Pandey, 2017). Additionally, the transformation of rural landscapes into a mosaic of rural and urban land use due to peri-urbanization results in the loss of green cover, vegetation, and wetlands, undermining the generation of vital ecosystem services (Morya & Punia, 2021).

The dynamic nature of ecosystem services value is evident in studies comparing their estimated worth before and after land use changes. Morya and Punia (2021) argue that prioritizing the conservation and enrichment of ecosystems of critical value is essential for sustainable urban development, shedding light on the importance of accounting for these shifts in urban planning. Moreover, the drainage of wetlands and natural features in peri-urban areas, a common consequence of urbanization, disrupts their flood regulation services, escalating flood risks in urban areas (Pandey, 2017).

These insights hold profound implications for urban planning in peri-urban regions. ### Research Questions, Hypotheses, and Rationale

**Research Question 1:** What is the temporal lag between declining EC and the altered supply of ES to urban areas?

**Hypothesis 1:** We hypothesize the existence of a temporal lag between the decline in EC and the corresponding changes in the supply of ES to urban areas. This lag is pivotal for comprehending the resilience of urban ecosystems and the potential consequences of delayed action in addressing ecological decline. Various authors have advanced several hypotheses concerning the nature of this temporal lag. Studies conducted by Chen et al. (2023) and Patru-Stupariu et al. (2020) underscore the importance of temporal analysis and how changes in EC affect the supply of ES to urban areas over time. Chen et al. (2023) emphasized the need for a comprehensive, multifaceted approach to quantifying ES indicators, indicating that EC dynamics can be influenced by a variety of factors, including urbanization. Patru-Stupariu et al. (2020) highlighted the significance of mapping as a foundation for spatial analyses, enabling the assessment of changes in mountain land use and its effects on ES. Both studies partly support the hypothesis that a temporal lag exists between declining EC and changes in ES supply to urban areas, as they emphasize the importance of understanding the dynamics and interplay between these two elements over time (Chen et al., 2023; Patru-Stupariu et al., 2020).

**Research Question 2:** Does the surplus of ES supplied to urban areas correspond to a decline in ES supplied to peri-urban areas?

**Hypothesis 2:** When urban areas receive an abundance of ecosystem services (ES), there might be a corresponding decline in ES supplied to peri-urban regions. This concept is crucial for unraveling the intricate connections between urban and peri-urban ecosystems and exploring potential trade-offs between them. Githiora-Murimi et al. (2022) have brought to light the dynamic nature of ES flow along the rural-urban spectrum, influenced by a myriad of interconnected factors. In their study exploring the supply of ES in the Yala swamp, Kenya, active engagement with local stakeholders responsible for managing and utilizing natural resources in the wetland was a central component. While the study doesn't explicitly delve into ES surpluses in urban core, it underscores the essential task of assessing ecosystems' capacity to furnish ES to diverse regions. The emphasis on involving stakeholders and gaining insights into ES distribution aligns with the hypothesis that trade-offs might indeed exist between ES supplied to urban and peri-urban areas (Githiora-Murimi et al., 2022).


While the literature acknowledges the ES>EC phenomenon, there exists a conspicuous dearth of comprehensive studies that delve into the temporal nuances and reciprocal consequences of this ecological deficit. This research endeavors to bridge this critical gap by meticulously dissecting land-cover changes, EC, and ES provisioning within the unique context of Delhi NCR, where urban expansion looms as an escalating concern (Barton et al., 2017; Feddern et al., 2023).

A review of the literature clearly illustrates the significantly high proportion of studies in the field concerned with the challenges of urbanization and ecosystem health. By focusing on peri-urban regions, where the ES>EC dynamic is particularly pronounced, this study addresses a critical knowledge gap in the existing body of literature (Jonas et al., 2015). The inclusion of peri-urban areas in this research is vital, as they often represent the frontlines of urban expansion and bear the brunt of ecosystem service deficits.

Understanding the temporal lag between EC decline and the altered supply of ES (RQ1) is pivotal for proactive policymaking. It enables the identification of a crucial intervention window between the decline in EC and the manifestation of adverse effects on ES supply, empowering policymakers with invaluable insights to implement timely measures that fortify ecosystem health and ensure the sustainable provisioning of ES (Ratcliffe et al., 2016). Additionally, exploring the reciprocal relationship between urban and peri-urban areas (RQ2) is pivotal for holistic urban planning. This exploration elucidates how the surplus of ES in urban cores can inadvertently impact peri-urban ecosystems, shedding light on potential trade-offs between them (Vogel et al., 2021).

```python
#!/usr/bin/env python
# coding: utf-8

# In[2]:


import ee
import geemap
import datetime

# Initialize Earth Engine
ee.Initialize()

# Define the FAO/GAUL administrative units dataset
dataset = ee.FeatureCollection('FAO/GAUL/2015/level2')

# Define a region of interest (ROI) for India
india_roi = ee.Geometry.Polygon(
    [[[
      68.11137880340998, 7.96553484576015],
      [88.3927306205192, 7.96553484576015],
      [88.3927306205192, 35.67459030295267],
      [68.11137880340998, 35.67459030295267],
      [68.11137880340998, 7.96553484576015]
      ]]
)

# Filter the administrative units dataset to select India
india_dataset = dataset.filterBounds(india_roi)

# Define a region of interest (ROI) for New Delhi
delhi_roi = ee.Geometry.Point(77.216721, 28.613939)

# Filter the India dataset to select New Delhi
delhi_feature = india_dataset.filterBounds(delhi_roi)

# Create a Map
Map = geemap.Map()
Map.centerObject(india_roi, 5)  # Center the map on India

# Add a layer for New Delhi
Map.addLayer(delhi_feature, {'color': 'FF0000'}, 'New Delhi')

# Buffer New Delhi and filter surrounding administrative units
buffered_delhi = delhi_feature.geometry().buffer(
    0.1)  # Adjust the buffer distance as needed
surrounding_admin_units = india_dataset.filterBounds(buffered_delhi)

# Add a layer for surrounding administrative units
Map.addLayer(surrounding_admin_units, {
             'color': 'Yellow'}, 'Surrounding Administrative Units')

# Function to calculate NDVI


def calculate_ndvi(image):
    ndvi = image.normalizedDifference(['B5', 'B4']).rename('NDVI')
    return image.addBands(ndvi)

# Function to get Landsat imagery for a given region


def get_landsat_imagery(region, start_date, end_date):
    landsat = ee.ImageCollection("LANDSAT/LC08/C01/T1_SR")         .filterBounds(
        region)         .filterDate(start_date, end_date)         .sort("CLOUD_COVER")         .mosaic()
    return landsat


# Define the time range for Landsat imagery
start_date = datetime.datetime(2020, 1, 1)
end_date = datetime.datetime(2021, 12, 31)

# Get Landsat imagery for New Delhi and the surrounding administrative units
landsat_region = surrounding_admin_units.geometry().union(buffered_delhi)
landsat_imagery = get_landsat_imagery(landsat_region, start_date, end_date)

# Calculate NDVI for New Delhi
delhi_with_ndvi = calculate_ndvi(
    landsat_imagery.clip(delhi_feature.geometry()))

# Calculate NDVI for surrounding administrative units
surrounding_admin_with_ndvi = calculate_ndvi(landsat_imagery)

# Add the NDVI layers to the map
Map.addLayer(delhi_with_ndvi.select('NDVI'), {
             'min': -1, 'max': 1, 'palette': ['blue', 'white', 'green']}, 'NDVI New Delhi')
Map.addLayer(surrounding_admin_with_ndvi.select('NDVI'), {
             'min': -1, 'max': 1, 'palette': ['blue', 'white', 'green']}, 'NDVI Surrounding Administrative Units')

# Combine the NDVI images for Delhi and surrounding administrative units
combined_ndvi = delhi_with_ndvi.add(surrounding_admin_with_ndvi)

# Add the combined NDVI layer to the map
Map.addLayer(combined_ndvi.select('NDVI'), {
             'min': -1, 'max': 1, 'palette': ['blue', 'white', 'green']}, 'NDVI Combined')

# Add a layer control to the map to toggle between layers
Map.addLayerControl()

# Display the Map
Map
```

##### Calculating Ecosystem Condition Indicators
Equation 1: Ecosystem Condition Indicator (ECI) Formula:

$$
ECI = \frac{V - VL}{VH - VL}
$$
Where:

ECI represents the Ecosystem Condition Indicator.<br>
`V` denotes the measured or observed value of the variable.<br>
`VH` stands for the high condition value for the variable (upper reference level).<br>
`VL` represents the low condition value (lower reference level).

```python
#!/usr/bin/env python
# coding: utf-8

# In[ ]:


import ee
import geemap
import datetime

# Initialize Earth Engine
ee.Initialize()

# Define the FAO/GAUL administrative units dataset
dataset = ee.FeatureCollection('FAO/GAUL/2015/level2')

# Define a region of interest (ROI) for India
india_roi = ee.Geometry.Polygon(
    [[[
      68.11137880340998, 7.96553484576015],
      [88.3927306205192, 7.96553484576015],
      [88.3927306205192, 35.67459030295267],
      [68.11137880340998, 35.67459030295267],
      [68.11137880340998, 7.96553484576015]
      ]]
)

# Filter the administrative units dataset to select India
india_dataset = dataset.filterBounds(india_roi)

# Define a region of interest (ROI) for New Delhi
delhi_roi = ee.Geometry.Point(77.216721, 28.613939)

# Filter the India dataset to select New Delhi
delhi_feature = india_dataset.filterBounds(delhi_roi)

# Create a Map
Map = geemap.Map()
Map.centerObject(india_roi, 5)  # Center the map on India

# Add a layer for New Delhi
Map.addLayer(delhi_feature, {'color': 'FF0000'}, 'New Delhi')

# Buffer New Delhi and filter surrounding administrative units
buffered_delhi = delhi_feature.geometry().buffer(
    0.1)  # Adjust the buffer distance as needed
surrounding_admin_units = india_dataset.filterBounds(buffered_delhi)

# Add a layer for surrounding administrative units
Map.addLayer(surrounding_admin_units, {
             'color': 'Yellow'}, 'Surrounding Administrative Units')

# Function to calculate NDVI


def calculate_ndvi(image):
    ndvi = image.normalizedDifference(['B4', 'B3']).rename('NDVI')
    return image.addBands(ndvi)

# Function to get Landsat imagery for a given region


def get_landsat_imagery(region, start_date, end_date, sensor):
    if sensor == 'L5':
        landsat = ee.ImageCollection("LANDSAT/LT05/C01/T1_SR")             .filterBounds(
            region)             .filterDate(start_date, end_date)             .sort("CLOUD_COVER")             .mosaic()
    elif sensor == 'L8':
        landsat = ee.ImageCollection("LANDSAT/LC08/C01/T1_SR")             .filterBounds(
            region)             .filterDate(start_date, end_date)             .sort("CLOUD_COVER")             .mosaic()
    return landsat


# Define the time range for Landsat imagery
years = [2001, 2005, 2010, 2013, 2015, 2020]
ndvi_layers = []

for year in years:
    start_date = datetime.datetime(year, 1, 1)
    end_date = datetime.datetime(year, 12, 31)

    # Determine the Landsat sensor based on the year
    sensor = 'L5' if year <= 2013 else 'L8'

    # Get Landsat imagery for New Delhi and the surrounding administrative units
    landsat_region = surrounding_admin_units.geometry().union(buffered_delhi)
    landsat_imagery = get_landsat_imagery(
        landsat_region, start_date, end_date, sensor)

    # Calculate NDVI for New Delhi
    delhi_with_ndvi = calculate_ndvi(
        landsat_imagery.clip(delhi_feature.geometry()))

    # Add the NDVI layer to the map
    Map.addLayer(delhi_with_ndvi.select('NDVI'), {
                 'min': -1, 'max': 1, 'palette': ['blue', 'white', 'green']}, f'NDVI New Delhi {year}')

    # Add the NDVI layer to the list for export
    ndvi_layers.append(delhi_with_ndvi.select('NDVI'))

# Combine the NDVI images for Delhi and surrounding administrative units
combined_ndvi = ee.ImageCollection(
    ndvi_layers).mosaic().rename('NDVI_Combined')

# Add the combined NDVI layer to the map
Map.addLayer(combined_ndvi.select('NDVI_Combined'), {
             'min': -1, 'max': 1, 'palette': ['blue', 'white', 'green']}, 'NDVI Combined')

# Add a layer control to the map to toggle between layers
Map.addLayerControl()

# Display the Map
Map
```
